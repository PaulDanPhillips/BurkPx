```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Week1
* Class Imbalance
        * df.Status.value_counts()
            * Negative    647
            * Melioid      48
    * Imbalanced roc_auc = 647/(647+48) =0.93
### Enet

```python
if not os.path.isdir('Figures'):
    os.makedirs('Figures')
os.chdir('Figures')

Enet = pd.read_csv('../Week1_EnetBurkPx_GridSearch.txt', sep='\t', index_col=0)
Enet['params2'] = Enet['params'].astype(str)
fig, ax = plt.subplots()
sns.boxplot(data=Enet, y= 'mean_test_score', x='params2', ax=ax)
plt.xticks(rotation=90)
fig.savefig('W1_Enet.png', dpi=300, bbox_inches="tight", transparent=False)
```

```python
{'C': 100, 'l1_ratio': 0.9}
```

### Logit

```python
LR = pd.read_csv('../Week1_LRBurkPx_GridSearch.txt', sep='\t', index_col=0)
LR['params2'] = LR['params'].astype(str)
fig, ax = plt.subplots()
sns.boxplot(data=LR, y= 'mean_test_score', x='params2', ax=ax)
plt.xticks(rotation=90)
fig.savefig('W1_LR.png', dpi=300, bbox_inches="tight", transparent=False)
```

```python
### Strong penalization
{'C': 0.01291549665014884, 'penalty': 'l1'}
{'C': 0.01291549665014884, 'penalty': 'l2'}
### Weak penalization
{'C': 100, 'penalty': 'l1'}
{'C': 100, 'penalty': 'l2'}
```
