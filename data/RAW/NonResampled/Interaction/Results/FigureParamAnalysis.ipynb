```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Interaction
* Class Imbalance
### Enet

```python
if not os.path.isdir('Figures'):
    os.makedirs('Figures')
os.chdir('Figures')

Enet = pd.read_csv('../Interaction_EnetBurkPx_GridSearch.txt', sep='\t', index_col=0)
Enet['params2'] = Enet['params'].astype(str)
fig, ax = plt.subplots()
sns.boxplot(data=Enet, y= 'mean_test_score', x='params2', ax=ax)
plt.xticks(rotation=90)
fig.savefig('W1_Enet.png', dpi=300, bbox_inches="tight", transparent=False)
```
