```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
```

```python
def paramGrab(w1path, w2path, w3path, w4path, timegroup1, timegroup2, timegroup3, timegroup4, params, model, assay):
    w1files=glob.glob(w1path)
    dflst=[]
    for file in w1files:
        df= pd.read_csv(file, sep='\t').assign(TimeGroup=timegroup1)
        dflst.append(df)
    w1_igg=pd.concat(dflst)
    w1_igg = w1_igg.loc[(w1_igg['params'] ==params) ].assign(Model=model) 
    
    #### Week2

    w2files=glob.glob(w2path)
    dflst=[]
    for file in w2files:
        df= pd.read_csv(file, sep='\t').assign(TimeGroup=timegroup2)
        dflst.append(df)
    w2_igg=pd.concat(dflst)
    w2_igg = w2_igg.loc[(w2_igg['params'] ==params) ].assign(Model=model) 

        #### Week3

    w3files=glob.glob(w3path)
    dflst=[]
    for file in w3files:
        df= pd.read_csv(file, sep='\t').assign(TimeGroup=timegroup3)
        dflst.append(df)
    w3_igg=pd.concat(dflst)
    w3_igg = w3_igg.loc[(w3_igg['params'] ==params) ].assign(Model=model)


    #### Week4

    w4files=glob.glob(w4path)
    dflst=[]
    for file in w4files:
        df= pd.read_csv(file, sep='\t').assign(TimeGroup=timegroup4)
        dflst.append(df)
    w4_igg=pd.concat(dflst)
    w4_igg = w4_igg.loc[(w4_igg['params'] ==params) ].assign(Model=model)

    DF = pd.concat([w1_igg, w2_igg, w3_igg, w4_igg]).assign(Assay=assay)
    return DF
```
